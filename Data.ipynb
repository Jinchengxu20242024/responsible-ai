```python
import pandas as pd
import numpy as np
from aequitas.group import Group
from aequitas.bias import Bias
from aequitas.plotting import Plot
```

```python
emotion_df = pd.read_csv("./Dataset/World_Happiness_Report_2024.csv")
emotion_df.head()
```

```python
emotion_df.drop_duplicates()
```

```python
# the most recent data
emotion_2023 = emotion_df[emotion_df["year"] == 2023]
emotion_2023
```

```python
# filter out nan Generosity and strip strings
emotion_2023 = emotion_2023[emotion_2023['Generosity'].notnull()]
emotion_2023["Country name"] = emotion_2023["Country name"].str.strip()
emotion_2023
```

```python
# Define income groups based on Log GDP per capita
def categorize_gdp(gdp):
    if gdp < 9:         # Threshold for low-income
        return 'low'
    elif gdp < 10:       # Threshold for middle-income
        return 'middle'
    else:
        return 'high'    # High-income

emotion_2023['income_group'] = emotion_2023['Log GDP per capita'].apply(categorize_gdp)
emotion_2023
```

```python
print("Min life ladder: ", min(emotion_2023['Life Ladder']))
print("Max life ladder: ", max(emotion_2023['Life Ladder']))
# Set a threshold (median of Life Ladder) to generate a synthetic score and label
threshold = emotion_2023['Life Ladder'].median()
# Create a 'score' column based on Life Ladder - this is our synthetic "prediction"
emotion_2023['score'] = (emotion_2023['Life Ladder'] >= threshold).astype(int)

# Create a 'label_value' column, which represents the ground truth (using the same logic for demo)
emotion_2023['label_value'] = (emotion_2023['Life Ladder'] >= threshold).astype(int)
```

```python
# Prepare columns for Aequitas
aequitas_df = emotion_2023[['score', 'label_value', 'income_group']]
aequitas_df
```

```python
# Initialize the Aequitas Group object
group = Group()

# Get group fairness metrics
xtab, _ = group.get_crosstabs(aequitas_df)

# Display fairness metrics for all income groups
group_metrics = group.list_groups(xtab)
group_metrics
```
